```python
import numpy as np
from numpy import log, sqrt
import time
from matplotlib import pyplot as plt
```

```python
# TASK 1
from math import pi
x = pi/5
s = 0
P = [0, pi/6, pi/4, pi/3]
for j in range(len(P)):
    l = 1
    for k in range(len(P)):
        if k != j:
            l *= (x - P[k]) / (P[j] - P[k])
#     print(l)
    s += np.abs(l)
print(s)
```

```python
# TASK 2
G = np.array([[1,1/2,1/3],
             [1/2,1/3,1/4],
             [1/3,1/4,1/5]])
b = np.array([0.4597, 0.3012, 0.2232])
np.linalg.solve(G, b)
def fsin(x):
    return -0.0099 + x * 1.1052 - x**2 * 0.249
X = np.linspace(0, 1, 10)
plt.plot(X, np.sin(X))
plt.plot(X, fsin(X))
```

```python
# TASK 3
def f(x):
    return x**(1/3)

def approx_f(x):
    return 1/(3*sqrt(3))+x

X = np.linspace(0,1.1,100)
plt.plot(X, f(X))
plt.plot(X, approx_f(X))
plt.legend(['$f$', 'Approx'])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Chebyshev alternance approximation')
plt.show()
```

```python
def MSI(x, l=1, iterations=6):
    for _ in range(iterations):
        x = l * (np.exp(2 * x) - 1)
    return x

print(MSI(-0.8))
```

```python
# TASK 5
from scipy.optimize import newton

def f(x):
    return np.log(x+2)-x*x

x = -0.6
x0 = newton(f, x, tol=1e-6, full_output=True)
print('Iterations: ', x0[1].iterations)
```
