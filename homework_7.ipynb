```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import lagrange, CubicSpline
```

```python
# def interpolation(x, y):
#     n = len(x)
#     coef = np.zeros((n, n))
#     coef[0,:] = y
#     step = n
#     for i in range(1, n):
#         step -= 1
#         for j in range(step):
#             coef[i, j] = (coef[i - 1, j] + coef[i - 1, j + 1]) / (x[j] - x[i - 1])
            
#     print(coef)

n = 8
a = 4
radius = 3 * a

def f(x):
    return 1 / (x * x + a * a)

true_x = np.linspace(-radius, radius, 100)
true_y = f(true_x)

x = np.linspace(-radius, radius, n)
y = f(x)

# interpolation(x, y)
L = lagrange(x, y)(true_x)
S = CubicSpline(x, y)(true_x)

plt.figure(figsize=(15,10))
plt.plot(true_x, true_y)
plt.plot(true_x, L)
plt.plot(true_x, S)
plt.title('Interpolations comparation')
plt.legend(['$f(x)$', 'Lagrange', 'Spline'])
plt.xlabel('$x$')
plt.ylabel('$y$')
```
